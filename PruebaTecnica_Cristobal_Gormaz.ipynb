```python
# @title
# ==========================================
# CONFIGURACIÓN DEL ENTORNO (EJECUTAR PRIMERO)
# ==========================================
import pandas as pd
import sqlite3

# 1) URLs de archivos crudos (Raw) en GitHub
urls = {
    'clientes': "https://raw.githubusercontent.com/vbroosing/prueba-tecnica/refs/heads/main/clientes.csv",
    'ventas': "https://raw.githubusercontent.com/vbroosing/prueba-tecnica/refs/heads/main/ventas.csv",
    'trabajadores': "https://raw.githubusercontent.com/vbroosing/prueba-tecnica/refs/heads/main/trabajadores.csv",
    'cargos': "https://raw.githubusercontent.com/vbroosing/prueba-tecnica/refs/heads/main/cargos.csv",
    'areas': "https://raw.githubusercontent.com/vbroosing/prueba-tecnica/refs/heads/main/areas.csv"
}

print("Preparando base de datos 'techstore.db'.. .")

# 2) Crear la conexión y la base de datos
conn = sqlite3.connect('techstore.db')

# 3) Leer cada CSV e insertarlo como tabla en la base de datos
for tabla, url in urls.items():
    try:
        df = pd.read_csv(url, sep=None, engine='python')
        df.to_sql(tabla, conn, index=False, if_exists='replace')
        print(f"Tabla '{tabla}' cargada exitosamente.")
    except Exception as e:
        print(f"Error al cargar la tabla '{tabla}': {e}")

conn.close()
print("¡Entorno listo! La base de datos relacional está disponible para ser consultada.")
```

```python
import pandas as pd
import re

productos = []

for _, fila in ventas_df.iterrows():

    venta_id = fila['id']

    items = fila['descripcion'].split(',')

    for item in items:

        item = item.strip()

        match = re.match(r'(\d+)x\s+(.*)', item)

        if match:

            cantidad = int(match.group(1))
            producto = match.group(2)

            productos.append([
                venta_id,
                producto,
                cantidad
            ])

productos_df = pd.DataFrame(
    productos,
    columns=[
        'id_venta',
        'producto',
        'unidades'
    ]
)

top_productos = (
    productos_df
    .groupby('producto', as_index=False)
    ['unidades']
    .sum()
    .sort_values('unidades', ascending=False)
    .head(10)
)

top_productos
```

###### Productos más vendidos

Se procesó la columna descripcion para separar las cantidades y nombres de productos contenidos en cada venta.

Posteriormente se consolidaron las unidades vendidas por producto y se obtuvo el ranking de los 10 productos con mayor volumen de ventas.

Este análisis permite identificar los productos con mayor rotación dentro del catálogo.

Top Clientes

Se identificaron los 5 clientes con mayor gasto histórico mediante JOIN entre clientes y ventas, agrupando por cliente y sumando total_dte.

```python
ventas_ret = ventas_df[['id_cliente','dte']].copy()

ventas_ret = ventas_ret.sort_values(
    ['id_cliente','dte']
)
```

```python
ventas_ret['siguiente_compra'] = (
    ventas_ret.groupby('id_cliente')['dte']
    .shift(-1)
)
```

```python
ventas_ret['diferencia_dte'] = (
    ventas_ret['siguiente_compra']
    - ventas_ret['dte']
)

ventas_ret.head(20)
```

```python
clientes_con_retencion = (
    ventas_ret[ventas_ret['diferencia_dte'] <= 200]
    ['id_cliente']
    .nunique()
)

total_clientes = ventas_ret['id_cliente'].nunique()

porcentaje_retencion = (
    clientes_con_retencion
    / total_clientes
    * 100
)

print("Clientes con recompra:", clientes_con_retencion)
print("Total clientes:", total_clientes)
print("Retención:", round(porcentaje_retencion, 2), "%")
```

**Se consideró retenido un cliente cuando al menos una compra fue seguida por otra compra dentro de los 200 DTE siguientes.**
